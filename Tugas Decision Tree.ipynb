```python
import pandas as pd
```

```python
df = pd.read_csv("Penjualan Tiket Bioskop.csv", delimiter=';')
df.head()
```

```python
inputs = df.drop('Purchased', axis='columns')
```

```python
target = df['Purchased']
```

```python
from sklearn.preprocessing import LabelEncoder
le_film = LabelEncoder()
le_genre = LabelEncoder()
le_gender = LabelEncoder()
```

```python
inputs['film_n'] = le_film.fit_transform(inputs['Film'])
inputs['genre_n'] = le_genre.fit_transform(inputs['Genre'])
inputs['gender_n'] = le_gender.fit_transform(inputs['Gender'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Film', 'Genre', 'Gender'], axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```

```python
example_data = [[le_film.transform(['Avengers'])[0], le_genre.transform(['Action'])[0], 23, le_gender.transform(['Male'])[0], 12.5]]
prediction = model.predict(example_data)

print(f"Prediction for Avengers, Action, 23, Male, 12.5: {prediction}")
```

```python
example_data_2 = [[le_film.transform(['Frozen 2'])[0], le_genre.transform(['Animation'])[0], 30, le_gender.transform(['Female'])[0], 10.0]]
prediction_2 = model.predict(example_data_2)

print(f"Prediction for Frozen 2, Animation, 30, Female, 10.0: {prediction_2}")
```
